## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([0, 1], [0, 3]) # Rysuję wykres
```

```python
plt.plot([0, 3, 4], [1, 2, 3], color='orange') # Rysuję wykres w wybranym kolorze
```

```python
hours = np.arange(7, 24)
temp = [ 6,   7,  9, 12, 13, 14, 16, 17, 17, 17, 17, 16, 14, 12, 11,  9,  8 ]
wind = [ 18, 18, 19, 21, 23, 24, 24, 23, 23, 21, 19, 18, 14, 13, 11, 11, 11 ]
```

```python
plt.xlabel('Godzina') # Dodaję opis osi
plt.ylabel('Temperatura')
plt.title('Prognoza temperatury i siły wiatru 10 kwietnia') # Dodaję tytuł

plt.plot(hours, temp, label='temperatura', color='red') # Dodaję opis dla wykresu
plt.plot(hours, wind, label='siła wiatru', color='lightblue')

plt.legend() # Wyświetlam legendę
```

```python
plt.subplot() # Tworzę siatkę

plt.xlabel('Godzina')
plt.ylabel('Temperatura/Siła wiatru')
plt.title('Prognoza temperatury i siły wiatru 10 kwietnia')

plt.plot(hours, temp, label='temperatura', color='red')
plt.plot(hours, wind, 'r--', label='siła wiatru', color='lightblue') # Zmieniam styl linii

plt.legend()
```

```python
fig, ax = plt.subplots()

x = np.arange(len(hours))
width = .25

temps = ax.bar(x - width / 2, temp, width, label="Temperatura") # Tworzę wykres składający się z dwóch kolumn
winds = ax.bar(x + width / 2, wind, width, label="Siła wiatru")

ax.set_ylabel('Temperatura/Siła wiatru')
ax.set_title('Prognoza temperatury i siły wiatru 10 kwietnia')
ax.set_xticks(x)
ax.set_xticklabels(hours)
ax.legend()

fig.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots()

ax.stackplot(hours, [ temp, wind ]) # Tworzę stackowany wykres
ax.set_title('Siła pogody')
ax.set_ylabel('Siła zjawisk atmosferycznych')

plt.show()
```

 ## Plotly

```python
import plotly.graph_objects as go
import pandas as pd
```

```python
skiprows = [x for x in range(0, 254)]
skiprows.append(283)
skiprows.append(283)
skiprows.append(284)
skiprows.append(285)
skiprows.remove(3)

uk_export = pd.read_excel('tradequarterlyq320final.xlsx', sheet_name="1 Total Trade by Country - A", skiprows=(0, 1, 2, 4, 5, 35, 36, 37), usecols="B:W", nrows=242) # Odczytujemy eksport
uk_import = pd.read_excel('tradequarterlyq320final.xlsx', sheet_name="1 Total Trade by Country - A", skiprows=skiprows, usecols="B:W", nrows=242) # Odczytujemy import

uk_import.rename(columns={'Unnamed: 1': 'Country'}, inplace=True)
uk_export.rename(columns={'Unnamed: 1': 'Country'}, inplace=True)
```

```python
sum = uk_import[2019].sum()

copy = uk_import.copy()

# Import w 2019 roku wyrażony w procentach
copy['Part'] = copy[2019] / sum * 100
copy.loc[copy['Part'] < .5, 'Country'] = 'Inne kraje'

figure = go.Figure( 
    data=go.Pie(values=copy['Part'], labels=copy['Country'], ), # Tworzę wykres kołowy
    layout=go.Layout(title='Import do Wielkiej Brytanii w 2019 roku') # Dodaję opis wykresu
)
figure.show()
```

```python
uk_import.set_index('Country', inplace=True)
uk_export.set_index('Country', inplace=True)
```

```python
import_poland = uk_import.loc['Poland']
export_poland = uk_export.loc['Poland']
```

```python
figure = go.Figure( 
    data=[go.Scatter(name="Import z Polski", y=import_poland, x=uk_import.columns), # Tworzę wykres liniowy
          go.Scatter(name="Eksport do Polski", y=export_poland, x=uk_export.columns)],
    layout=go.Layout(title="Wielkość importu i eksportu w handlu Polski z Wielką Brytanią"), 
)
figure.show()
```

```python
figure.write_html('trade_uk_pl.html') # Zapisuję plik
```

```python
trade_poland = pd.DataFrame({ 'Import': import_poland, 'Export': export_poland})
trade_poland['ImportToExportRatio'] = import_poland / (import_poland + export_poland)
trade_poland['ExportToImportRatio'] = export_poland / (import_poland + export_poland)

trade_poland.loc[2019, ['ImportToExportRatio', 'ExportToImportRatio']]

figure = go.Figure( 
    data=[
          go.Bar(x=trade_poland.index, y=trade_poland['ImportToExportRatio'], name="Import z Polski"), # Tworzę wykres słupkowy
          go.Bar(x=trade_poland.index, y=trade_poland['ExportToImportRatio'], name="Export z Polski"),
    ],
    layout=go.Layout(title="Porównanie importu i eksportu w handlu Polski z Wielką Brytanią"), 
)
figure.show()
```

## Plotly Express

```python
import plotly.express as px
```

```python
fig = px.pie(copy, values='Part', names='Country', title='Import do Wielkiej Brytanii w 2019 roku') # Rysuję wykres kołowy
fig.show()
```

```python
trade_poland.reset_index(inplace=True)
trade_poland.rename(columns={'index': 'Year'}, inplace=True)
```

```python
copy = trade_poland.copy()

import_poland = copy[["Year", "ImportToExportRatio", "Import"]]
export_poland = copy[["Year", "ExportToImportRatio", "Export"]]

import_poland.insert(2, "Operation", "Import", True)
export_poland.insert(2, "Operation", "Export", True)

trade_poland = import_poland.rename(columns={'ImportToExportRatio': 'Ratio','Import': 'Value'}).append(export_poland.rename(columns={'ExportToImportRatio': 'Ratio', 'Export': 'Value' }))
```

```python
fig = px.bar(trade_poland, y="Ratio", x="Year", color="Operation", barmode='group', title="Porównanie importu i eksportu w handlu Polski z Wielką Brytanią") # Rysuję wykres słupkowy
fig.show()
```

```python
fig = px.line(trade_poland, y="Value", x="Year", color="Operation", title="Wielkość importu i eksportu w handlu Polski z Wielką Brytanią") # Rysuję wykres liniowy
fig.show()
```

## Seaborn

```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_excel('tradequarterlyq320final.xlsx', sheet_name="1 Total Trade by Country - A", skiprows=(0, 1, 2, 4, 5, 35, 36, 37), usecols="B:W", nrows=242).rename(columns={'Unnamed: 1': 'Country'}).set_index('Country')
data
```

```python
data = data.sort_values(by=2019, ascending=False)
```

```python
sns.catplot(data=data, y=2019) # Rysuję wykres rozłożenia importu z Wielkiej Brytanii przez różne kraje na świecie
```

```python
# Tworzę wykres eksportu produktów z Wielkiej Brytanii do 5 krajów o największym ich imporcie w 2019 roku, na podstawie danych z lat od 2015 do 2019
# Zmieniam aspect ratio
sns.relplot(data=data.head(20)[[x for x in range(2015, 2020)]], aspect=4)
```
